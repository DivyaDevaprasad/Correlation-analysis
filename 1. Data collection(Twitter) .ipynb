```python
def scrape_twitter(comp):
    c = twint.Config()
    c.Search = comp
        
    c.Format = "Username: {username} | Date:{date} | Tweet: {tweet}" # Custom output format
    
    c.Pandas = True
    c.Since = '2019-12-01'
    c.Until = '2020-01-01'
    c.Lang = 'en'

    twint.run.Search(c)

    # Data Collection into dataframe
    df = twint.storage.panda.Tweets_df

    # Save the tweets as csv
    df.to_csv(comp+"_tweets.csv")

    
companies = {'WMT','XOM','BRK', 'AMZN', 'AAPL'}

for comp in companies:
    ''' 
        For each company, scrape twitter data
    '''
    scrape_twitter(comp)
    
```
